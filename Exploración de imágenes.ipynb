# Exploración de imágenes

## MNIST

El conjunto de datos MNIST es una gran colección de dígitos escritos a mano que se utiliza comúnmente para entrenar sistemas de procesamiento de imágenes, particularmente para clasificación.

Consta de 60.000 imágenes de entrenamiento y 10.000 imágenes de prueba, cada una de las cuales es una imagen en escala de grises de 28x28 píxeles.

MNIST puede descargarse de forma independiente o como [parte de `Keras`](https://keras.io/api/datasets/mnist/).



```python
from tensorflow import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

Extraemos una imagen, por ejemplo la primera (índice 0).

```python
img   = x_train[0]
label = y_train[0]

img
```

Mostramos la matriz correspondiente a la imagen usando `pandas`.

```python
import pandas as pd
pd.set_option('display.max_columns', None)

df_img = pd.DataFrame(img)
df_img
```

Mostramos la imagen como gráfico usando `matplotlib`.

```python
import matplotlib.pyplot as plt

plt.imshow(img, cmap='gray')
plt.title(f"Label: {label}")
plt.axis('off')
plt.show()
```

## CIFAR10

El conjunto de datos CIFAR10 es una colección de imágenes que se utiliza comúnmente para entrenar algoritmos de aprendizaje automático y visión artificial.

Consta de 60,000 imágenes en color de 32x32 píxeles, divididas en 10 clases diferentes, con 6,000 imágenes por clase. Las clases incluyen aviones, automóviles, pájaros, gatos, ciervos, perros, ranas, caballos, barcos y camiones.

CIFAR10 puede descargarse de forma independiente o como [parte de `Keras`](https://keras.io/api/datasets/cifar10/).

```python
import zipfile
with zipfile.ZipFile('cats_and_dogs_small.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('./cats_and_dogs_small/train/cats/cat.2.jpg')
plt.imshow(img)
plt.title('Cat')
plt.axis('off')
```

```python

dog_dir = './cats_and_dogs_small/train/dogs'
cat_dir = './cats_and_dogs_small/train/cats'

# Obtener 5 imágenes de perros y 5 imágenes de gatos
dog_images = [os.path.join(dog_dir, img) for img in os.listdir(dog_dir) if img.endswith('.jpg')][:5]
cat_images = [os.path.join(cat_dir, img) for img in os.listdir(cat_dir) if img.endswith('.jpg')][:5]

# Dibujar en una cuadrícula 5x2
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

# Mostrar imágenes de perros
for i, img_path in enumerate(dog_images):
    img = mpimg.imread(img_path)
    axes[0, i].imshow(img)
    axes[0, i].set_title('Dog')
    axes[0, i].axis('off')

# Mostrar imágenes de gatos
for i, img_path in enumerate(cat_images):
    img = mpimg.imread(img_path)
    axes[1, i].imshow(img)
    axes[1, i].set_title('Cat')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```
